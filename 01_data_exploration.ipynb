```python
import pandas as pd
from pathlib import Path


# If files are huge, change preview_rows to a small number
preview_rows = 5000

paths = {
    "crisis": "data/raw/crisis_dataset.csv",
    "drone": "data/raw/drone_disaster_area_identification_dataset.csv",
    "natural": "data/raw/natural_disasters_2024.csv",
}

# Load with a safe preview, then full load if small
dfs = {}
for key, p in paths.items():
    size_kb = Path(p).stat().st_size / 1024
    print(f"Loading {p} ({round(size_kb,1)} KB)...")
    if size_kb > 5_000:  # > ~5 MB, only preview
        print("  large file — reading first", preview_rows, "rows for preview")
        dfs[key] = pd.read_csv(p, nrows=preview_rows)
        dfs[key]._preview_only = True
    else:
        dfs[key] = pd.read_csv(p)
        dfs[key]._preview_only = False

# quick peek
for k, df in dfs.items():
    print(f"\n=== {k.upper()} (shape: {getattr(df,'shape',None)}) ===")
    display(df.head())
```

```python
import numpy as np

def try_parse_datetimes(df, sample=1000):
    parsed = []
    for col in df.columns:
        if df[col].dtype == object:
            sample_vals = df[col].dropna().astype(str).head(sample)
            if sample_vals.empty:
                continue
            # Heuristic: many values parse as dates
            parsed_count = 0
            for v in sample_vals:
                try:
                    pd.to_datetime(v)
                    parsed_count += 1
                except Exception:
                    pass
            if parsed_count / len(sample_vals) > 0.6:
                parsed.append(col)
    return parsed

for name, df in dfs.items():
    candidates = try_parse_datetimes(df)
    print(f"{name}: datetime candidates -> {candidates}")
    for c in candidates:
        try:
            dfs[name][c] = pd.to_datetime(dfs[name][c], errors="coerce")
            print(f"  parsed column {c} -> dtype {dfs[name][c].dtype}")
        except Exception as e:
            print("  failed to parse", c, e)
```

```python
import pandas as pd
def detect_parse_datetime(df):
    cand = []
    for c in df.columns:
        if df[c].dtype == object:
            try:
                parsed = pd.to_datetime(df[c], errors='coerce')
                if parsed.notna().sum() / max(1, len(parsed)) > 0.6:
                    cand.append(c)
                    df[c] = parsed
            except: 
                pass
    return cand

for name, df in dfs.items():
    cands = detect_parse_datetime(df)
    print(f"{name} parsed datetime candidates: {cands}")
```

```python
import pandas as pd
import numpy as np

def detect_latlon(df):
    lat_col, lon_col = None, None
    for col in df.columns:
        vals = pd.to_numeric(df[col], errors='coerce').dropna()
        if len(vals) == 0:
            continue
        if vals.between(-90, 90).mean() > 0.95 and lat_col is None:
            lat_col = col
        elif vals.between(-180, 180).mean() > 0.95 and lon_col is None:
            lon_col = col
    return lat_col, lon_col

for name, df in dfs.items():
    lat, lon = detect_latlon(df)
    print(f"\n{name.upper()}:")
    print("  Latitude column:", lat)
    print("  Longitude column:", lon)
    if lat and lon:
        print("  Range lat:", df[lat].min(), "→", df[lat].max())
        print("  Range lon:", df[lon].min(), "→", df[lon].max())
```

```python
# Define grid resolution (0.01° ≈ 1.1 km)
grid_res = 0.01

for name, df in dfs.items():
    lat, lon = detect_latlon(df)
    if lat and lon:
        df['__lat'] = pd.to_numeric(df[lat], errors='coerce')
        df['__lon'] = pd.to_numeric(df[lon], errors='coerce')
        df['grid_x'] = (df['__lon'] / grid_res).round().astype('Int64')
        df['grid_y'] = (df['__lat'] / grid_res).round().astype('Int64')
        print(f"{name}: created grid_x, grid_y for spatial mapping.")
```

```python
for name, df in dfs.items():
    # find datetime column
    time_cols = df.select_dtypes(include='datetime').columns.tolist()
    if len(time_cols) == 0 or '__lat' not in df.columns:
        print(f"{name}: missing datetime or lat/lon — skipping.")
        continue

    tcol = time_cols[0]
    df['time_bin'] = df[tcol].dt.floor('1D')  # daily bins

    agg = df.groupby(['time_bin', 'grid_x', 'grid_y']).size().reset_index(name='count')
    print(f"\n{name.upper()} aggregation preview:")
    display(agg.head())

    # Save for reuse
    agg.to_csv(f"data/processed/{name}_spatiotemporal.csv", index=False)
    print(f"Saved aggregated dataset: data/processed/{name}_spatiotemporal.csv")
```

```python
import pandas as pd
import glob

# Load all processed spatiotemporal datasets
files = glob.glob("data/processed/*_spatiotemporal.csv")
dfs_combined = [pd.read_csv(f) for f in files]

# Merge into one combined dataset
combined = pd.concat(dfs_combined, ignore_index=True)

# Ensure proper datetime dtype
combined['time_bin'] = pd.to_datetime(combined['time_bin'])

# Sort by time for consistency
combined.sort_values(['time_bin', 'grid_x', 'grid_y'], inplace=True)

# Save the combined dataset
combined.to_csv("data/processed/combined_disaster_data.csv", index=False)
print("✅ Combined dataset saved. Shape:", combined.shape)
display(combined.head())
```

```python
import numpy as np

def make_sequences(df, seq_len=7):
    X, y = [], []
    for grid, sub in df.groupby(['grid_x','grid_y']):
        sub = sub.sort_values('time_bin')
        vals = sub['count'].values
        if len(vals) > seq_len:
            for i in range(len(vals) - seq_len):
                X.append(vals[i:i+seq_len])
                y.append(vals[i+seq_len])
    return np.array(X), np.array(y)

X, y = make_sequences(combined)
print("X shape:", X.shape, "y shape:", y.shape)

# Save for reuse
np.save("data/processed/X_seq.npy", X)
np.save("data/processed/y_seq.npy", y)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load arrays
X = np.load("data/processed/X_seq.npy")
y = np.load("data/processed/y_seq.npy")

# Reshape for LSTM input: (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build simple LSTM model
model = models.Sequential([
    layers.Input(shape=(X.shape[1], 1)),
    layers.LSTM(64, activation='tanh'),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(X, y, epochs=20, batch_size=64, validation_split=0.2, verbose=1)
```
